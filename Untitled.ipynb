```python
# Data Science
import pandas as pd
import numpy as np

# Data Vizualization
import seaborn as sns
import matplotlib.pyplot as plt

# database
import sqlite3


import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, text, inspect, func
```

```python
# Create engine using the `spacex.sqlite` database file
engine = create_engine("sqlite:///../Resources/santa_barbara_food.sqlite")
```

```python
# Define the SQL query
conn = sqlite3.connect("sqlite:///../Resources/santa_barbara_food.sqlite")

query = """
SELECT categories, COUNT(*) as count
FROM santa_barbara_food
GROUP BY categories
ORDER BY count DESC;
"""
cursor = conn.cursor()
cursor.execute(query)
results = cursor.fetchall()

df = pd.read_sql(text(query), con=engine)
df.head()
```
